## Projet : Web scraping du site Cars.com
Nous recherchons le nom des voitures, leur prix, leur prix par mois, la note de la voiture , l'entreprise fournisseur et la ville.

```python
# Importer les librairies
from bs4 import BeautifulSoup
from  selenium import webdriver
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://www.cars.com/shopping/results/?stock_type=all&makes%5B%5D=mercedes_benz&models%5B%5D=&maximum_distance=30&zip=%27"
```

```python
def get_source_page(url):
    driver = webdriver.Chrome()
    driver.get(url)
    driver.implicitly_wait(30)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source,"html.parser")
    return soup
```

```python
def get_data(url):  
    """Cette fonction prend en paramètres l'url du site et renvoie un dataframe contenant
    les informations relatives à chaque voiture"""
    # Initialisation des lites devant recevoir les informations
    names = []
    prices = []
    price_bymonth = []
    notes = []
    companies = []
    countries = []
    
    # Récupérer le document HTML du site
    soup = get_source_page(url)
    # Récupérer tous les blocs de voitures
    elements = soup.find_all("div",class_="vehicle-card")
    for i in range(len(elements)):
        # Récupérer le nom de chaque voiture
        name_find = elements[i].find("h2",class_="title")
        if name_find is not None:
            names.append(name_find.text.strip())
        else :
            names.append("N/A")
            
        # Récupérer le prix de chaque voiture
        price_find = elements[i].find("span",class_="primary-price")
        if price_find is not None:
            prices.append(price_find.text.strip())
        else :
            prices.append("N/A")
 
        # Récupérer le prix par mois de chaque voiture
        price_bymonth_find = elements[i].find("span",class_="js-estimated-monthly-payment-formatted-value-with-abr")
        if price_bymonth_find is not None:
            price_bymonth.append(price_bymonth_find.text.strip())
        else :
            price_bymonth.append("N/A")

        # Récupérer la note de chaque voiture
        note_find = elements[i].find("span",class_="sds-rating__count")
        if note_find is not None:
            notes.append(note_find.text.strip())
        else :
            notes.append("N/A")
    
        # Récupérer le nom de la compagnie de chaque voiture
        company_find = elements[i].find("div",class_="dealer-name")
        if company_find is not None:
            companies.append(company_find.text.strip())
        else :
            companies.append("N/A")
         
        # Récupérer le nom de la ville de chaque voiture
        country_find = elements[i].find("div",class_="miles-from")
        if country_find is not None:
            countries.append(country_find.text.strip())
        else :
            countries.append("N/A")
        
        my_dict = {"Name":names,"Price":prices,"Price_bymonth":price_bymonth,"Notes":notes,"Dealer_company":companies,
                  "Countries":countries}

        df = pd.DataFrame(my_dict)
    return df
```

```python
data = get_data(url)
```

```python
def get_data_bypage(nb_pages):
    """Cette fonction prend en paramètres un nombre données de pages à investiguer 
    et renvoie un dataframe contenant les différentes informations"""
    # Initialisation d'un nouveau dataframe
    final_data = pd.DataFrame(columns= ["Name","Price","Price_bymonth","Notes","Dealer_company","Countries"])
    for p in range(1,nb_pages+1) :
        # Url type selon le numéro de la page
        url_type = f"https://www.cars.com/shopping/results/?makes[]=mercedes_benz&maximum_distance=30&models[]=&page={p}&stock_type=all&zip=%27"
        # Rechercher les informations sur la page
        data_find = get_data(url_type)
        # Ajouter les informations trouvées au dataframe initié au début
        final_data = pd.concat([final_data,data_find], axis=0)
        # Réinitialiser les index du dataframe
        final_data = final_data.reset_index(drop=True)
    return final_data
```
