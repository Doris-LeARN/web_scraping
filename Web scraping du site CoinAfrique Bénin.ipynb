## Projet : Web scraping du site CoinAfrique Bénin
Nous recherchons le nom des articles, leur prix, leur date de mis en vente , la ville et l'entreprise fournisseur

```python
from bs4 import BeautifulSoup
from  selenium import webdriver
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://bj.coinafrique.com/"
```

```python
def get_source_page(url):
    driver = webdriver.Chrome()
    driver.get(url)
    driver.implicitly_wait(15)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source,"html.parser")
    return soup
```

```python
def get_data(url):  

    names = []
    prices = []
    countries = []
    dates = []
    companies = []
    
    soup = get_source_page(url)
    elements = soup.find_all("div",class_="col s6 m4 l3")
    for i in range(len(elements)):
    
        name_find = elements[i].find("p",class_="ad__card-description")
        if name_find is not None:
            names.append(name_find.text)
        else :
            names.append("N/A")

        price_find = elements[i].find("p",class_="ad__card-price")
        if price_find is not None:
            prices.append(price_find.text)
        else :
            prices.append("N/A")

        country_find = elements[i].find("p",class_="ad__card-location")
        if country_find is not None:
            countries.append(country_find.text[11:])
        else :
            countries.append("N/A")

        date_find = elements[i].find("div",class_="card-title ad__card-timesince")
        if date_find is not None:
            dates.append(date_find.text)
        else :
            dates.append("N/A")

        company_find = elements[i].find("div",class_="card-action ad__card-action is-pro")
        if company_find is not None:
            companies.append(company_find.text)
        else :
            companies.append("N/A")

        my_dict = {"Name":names,"Price":prices,"Country":countries,"Date_added_to_sale":dates,"Supplier_company":companies}

        df = pd.DataFrame(my_dict)
    return df
```

```python
data = get_data(url)
```
